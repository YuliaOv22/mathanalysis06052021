**__6.__ Исследовать на экстремум функцию:**

### $U=x^2y+\frac{1}{3}y^3+2x^2+3y^2-1$

#### $\begin{cases}
   U'_{x}=2xy+4x=0, \\
   U'_{y}=x^2+y^2+6y=0 
 \end{cases}$

Получаются точки $(0, -6)$, $(0, 0)$, $(-2\sqrt{2}, -2)$, $(2\sqrt{2}, -2)$.

#### $U''_{xx}=2y+4$ 
#### $U''_{xy}=U''_{yx}=2x$  
#### $U''_{yy}=2y+6$
 
####  $$\begin{pmatrix}
2y+4 & 2x\\ 
2x & 2y+6\\
\end{pmatrix}$$

Главные миноры:

#### $\Delta_{1}=U''_{xx}=2y+4$

#### $\Delta_{2}=\begin{vmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\

#### $\Delta_{1}=U''_{xx}=-8<0$

#### $\Delta_{2}=\begin{vmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\
\end{vmatrix}=4(-6)^2+20(-6)+24-4*0^2=48>0$

Так как первый минор отрицательный, а следующий минор положительный, точка $(0, -6)$ является максимумом.

2. Миноры для точки $(0, 0)$:

#### $\Delta_{1}=U''_{xx}=4>0$

#### $\Delta_{2}=\begin{vmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\
\end{vmatrix}=24>0$

Так как оба минора положительные, точка $(0, 0)$ является минимумом.

3. Миноры для точек $(-2\sqrt{2}, -2)$ и $(2\sqrt{2}, -2)$:

#### $\Delta_{1}=U''_{xx}=0$

Первый минор равен $0$, следовательно теория Гессе не работает.

**Ответ:** 

${\color {red}{(0, -6) - максимум }}$

${\color {red}{(0, 0) - минимум}}$

```python
import numpy as np
np.set_printoptions(precision=2, suppress=True)

a = np.array([[4, 0], [0, 6]])
b = np.array([[-8, 0], [0, -6]])
w, v = np.linalg.eig(a)
q, p = np.linalg.eig(b)

print(f'Матрица Гёссе:\n{a}')
print(f'Собственные значения:\n{w}')

print(f'\nМатрица Гёссе:\n{b}')
print(f'Собственные значения:\n{q}')
```

**__8* *.__ С помощью метода наименьших квадратов (МНК) подобрать значения параметров $a$ и $b$ для сигмоидальной функции $y=\frac{1}{1+e^{-(ax+b)}}$ по заданному распределению. Построить график.**

Представим уравнение в неявном виде:
#### $y-\frac{1}{1+e^{-(ax+b)}}=0$

Возьмем правую часть в виде погрешности $\varepsilon$ для каждой точки:

#### $y_{i}-\frac{1}{1+e^{-(ax_{i}+b)}}=\varepsilon_{i}$

Составим функцию ошибок:

#### $U(a,b)=\varepsilon^2_{1}+\varepsilon^2_{2}+\varepsilon^2_{3}+\varepsilon^2_{4}+...+\varepsilon^2_{n}$

Найдем частные производные.

#### $\begin{cases}
   U'_{a}=\sum\limits_{i=1}^n(-\frac{2x_{i}\Bigr(y_{i}-\frac{1}{e^{-(ax_{i}+b)}+1}\Bigr)e^{-(ax_{i}+b)}}{(e^{-(ax_{i}+b)}+1)^2})=0, \\
   U'_{b}=\sum\limits_{i=1}^n(-\frac{2\Bigr(y_{i}-\frac{1}{e^{-(ax_{i}+b)}+1}\Bigr)e^{-(ax_{i}+b)}}{(e^{-(ax_{i}+b)}+1)^2})=0 
 \end{cases}$
 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
%matplotlib inline

# Задаем рандомное распределение точек
xi = np.random.uniform(-8, 8, 200)
yi = 1/(1 + np.exp(-np.random.uniform(0.5, 1.5, 200)*xi - np.random.uniform(-1, 1, 200)))
plt.plot(xi, yi, marker="o", ls="", c='purple')

# Численное решение системы уравнений
def equations(p):
    a, b = p
    return (sum(-(2*xi*(yi - 1/(np.exp(-a*xi-b)+1))*(np.exp(-a*xi-b)+1))/(np.exp(-a*xi-b)+1)**2), 
            sum(-(2*(yi - 1/(np.exp(-a*xi-b)+1))*(np.exp(-a*xi-b)+1))/(np.exp(-a*xi-b)+1)**2))

a, b =  fsolve(equations, (1, 1))
print (f'Коэффициент а: {a}\nКоэффициент b: {b}\n')

# Аппроксимация функции y = 1/(1 + np.exp(-a*x - b)) с помощью метода наименьших квадратов (МНК)
x = np.linspace(-8, 8, 200)
y = 1/(1 + np.exp(-a*x - b))
plt.plot(x, y, c='red')

plt.show()
```
