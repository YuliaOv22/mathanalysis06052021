**__1.__ Исследовать на условный экстремум функцию**

#### $U=3-8x+6y,$

если

#### $x^2+y^2=36$

По методу Лагранжа получаем:

$L(x, y, \lambda)=3-8x+6y+\lambda*(x^2+y^2-36)$

Найдем частные производные:

$\begin{cases}
   L'_{x}=-8+\lambda*2x=0, \\
   L'_{y}=6+\lambda*2y=0, \\
   L'_{\lambda}=x^2+y^2-36=0
 \end{cases}$

$\begin{cases}
   x=\frac{4}{\lambda}, \\
   y=-\frac{3}{\lambda}, \\
   (\frac{4}{\lambda})^2+(-\frac{3}{\lambda})^2-36=0
 \end{cases}
 \Rightarrow 
 \begin{cases}
   x=\frac{4}{\lambda}, \\
   y=-\frac{3}{\lambda}, \\
   \frac{25}{\lambda^2}=36
 \end{cases}
 \Rightarrow 
 \begin{cases}
   x=\frac{4}{\lambda}, \\
   y=-\frac{3}{\lambda}, \\
   \lambda=\pm\frac{5}{6}
 \end{cases}$

Получаем две точки: $(\frac{24}{5},-\frac{18}{5},\frac{5}{6})$ и $(-\frac{24}{5},\frac{18}{5},-\frac{5}{6})$

Найдем вторые производные:
$L''_{xx}=2\lambda$
$L''_{yy}=2\lambda$
$L''_{\lambda \lambda}=0$
$L''_{xy}=L''_{yx}=0$
$L''_{x\lambda}=L''_{\lambda x}=2x$
$L''_{y\lambda}=L''_{\lambda y}=2y$

Составим матрицу Гёссе:
$\begin{pmatrix}
L''_{\lambda \lambda} & L''_{\lambda x} & L''_{\lambda y}\\ 
L''_{x \lambda} & L''_{xx} & L''_{xy}\\
L''_{y \lambda} & L''_{yx} & L''_{yy}\\
\end{pmatrix}\Rightarrow
\begin{pmatrix}
0 & 2x & 2y\\ 
2x & 2\lambda & 0\\
2y & 0 & 2\lambda\\
\end{pmatrix}$

Найдем определитель в общем виде:

$\Delta=\begin{vmatrix}
0 & 2x & 2y\\ 
2x & 2\lambda & 0\\
2y & 0 & 2\lambda\\
\end{vmatrix}=
\begin{vmatrix}
+ & - & +\\ 
- & + & -\\
+ & - & +\\
\end{vmatrix}=0*
\begin{vmatrix}
2\lambda & 0\\
0 & 2\lambda\\
\end{vmatrix}-2x*
\begin{vmatrix}
2x & 0\\
2y & 2\lambda\\
\end{vmatrix}+2y*
\begin{vmatrix}
2x & 2\lambda\\
2y & 0\\
\end{vmatrix}=(-2x)* 4x\lambda+2y *(-4y\lambda)=-8\lambda(x^2+y^2)$

$\Delta=-8\lambda*36 = -288\lambda$

Таким образом, знак определителя зависит только от знака $\lambda$.

Если $\lambda=\frac{5}{6}$, то $\Delta<0$, следовательно $(\frac{24}{5},-\frac{18}{5},\frac{5}{6})$ - точка минимума.

Если $\lambda=-\frac{5}{6}$, то $\Delta>0$, следовательно $(-\frac{24}{5},\frac{18}{5},-\frac{5}{6})$ - точка максимума.

**Ответ:** 

${\color {red}{(\frac{24}{5},-\frac{18}{5},\frac{5}{6}) - минимум}}$

${\color {red}{(-\frac{24}{5},\frac{18}{5},-\frac{5}{6}) - максимум }}$

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
import numpy

fig = pylab.figure()
axes = Axes3D(fig)


def makeData1 ():
    x = numpy.arange (-6, 6, 0.0001)
    U = numpy.arange (140, 340, 20)
    xgrid, Ugrid = numpy.meshgrid(x, U)
    ygrid = numpy.sqrt(36-xgrid**2)
    return xgrid, ygrid, Ugrid
x, y, U = makeData1()
axes.plot_surface(x, y, U, color='blue')


def makeData2 ():
    x = numpy.arange (-6, 6, 0.0001)
    U = numpy.arange (140, 340, 20)
    xgrid, Ugrid = numpy.meshgrid(x, U)
    ygrid = -numpy.sqrt(36-xgrid**2)
    return xgrid, ygrid, Ugrid
x, y, U = makeData2()
axes.plot_surface(x, y, U, color='red')

def makeData3 ():
    x = numpy.arange (-10, 21, 1)
    y = numpy.arange (-10, 21, 1)
    xgrid, ygrid = numpy.meshgrid(x, y)
    Ugrid = -8*xgrid+6*ygrid+3
    return xgrid, ygrid, Ugrid
x, y, U = makeData3()
axes.plot_surface(x, y, U, color='purple')

pylab.show()
```

**__2.__ Исследовать на условный экстремум функцию**

$U=2x^2+12xy+32y^2+15,$

если

$x^2+16y^2=64$

По методу Лагранжа получаем:

$L(x, y, \lambda)=2x^2+12xy+32y^2+15+\lambda*(x^2+16y^2-64)$

Найдем частные производные:

$\begin{cases}
   L'_{x}=4x+12y+\lambda*2x=0, \\
   L'_{y}=12x+64y+\lambda*32y=0, \\
   L'_{\lambda}=x^2+16y^2-64=0
 \end{cases}$

 $\begin{cases}
   x=-\frac{6y}{2+\lambda}, \\
   y=-\frac{3x}{8(2+\lambda)}, \\
   (-\frac{6y}{2+\lambda})^2+16*(-\frac{3x}{8(2+\lambda)})^2-64=0
 \end{cases}
 \Rightarrow 
 \begin{cases}
   x=-\frac{6y}{2+\lambda}, \\
   y=-\frac{3x}{8(2+\lambda)}, \\
   \frac{36y^2}{(2+\lambda)^2}+\frac{9x^2}{4(2+\lambda)^2}=64
 \end{cases}$

Получаем четыре точки: 

$(-4\sqrt{2},-\sqrt{2},-\frac{7}{2})$ 

$(-4\sqrt{2},\sqrt{2},-\frac{1}{2})$ 

$(4\sqrt{2},-\sqrt{2},-\frac{1}{2})$ 

$(4\sqrt{2},\sqrt{2},-\frac{7}{2})$ 

Найдем вторые производные:
$L''_{xx}=4+2\lambda$
$L''_{yy}=64+32\lambda$
$L''_{\lambda \lambda}=0$
$L''_{xy}=L''_{yx}=12$
$L''_{x\lambda}=L''_{\lambda x}=2x$
$L''_{y\lambda}=L''_{\lambda y}=32y$

Составим матрицу Гёссе:
$\begin{pmatrix}
L''_{\lambda \lambda} & L''_{\lambda x} & L''_{\lambda y}\\ 
L''_{x \lambda} & L''_{xx} & L''_{xy}\\
L''_{y \lambda} & L''_{yx} & L''_{yy}\\
\end{pmatrix}\Rightarrow
\begin{pmatrix}
0 & 2x & 32y\\ 
2x & 4+2\lambda & 12\\
32y & 12 & 64+32\lambda\\
\end{pmatrix}$

Найдем определитель в общем виде:

$\Delta=\begin{vmatrix}
0 & 2x & 32y\\ 
2x & 4+2\lambda & 12\\
32y & 12 & 64+32\lambda\\
\end{vmatrix}=
\begin{vmatrix}
+ & - & +\\ 
- & + & -\\
+ & - & +\\
\end{vmatrix}=0*
\begin{vmatrix}
4+2\lambda & 12\\
12 & 64+32\lambda\\
\end{vmatrix}-2x*
\begin{vmatrix}
2x & 12\\
32y & 64+32\lambda\\
\end{vmatrix}+32y*
\begin{vmatrix}
2x & 4+2\lambda\\
32y & 12\\
\end{vmatrix}=(-2x)*(128x+64x\lambda-384y)+32y*(24x-128y-64y\lambda)=-256x^2-128x^2\lambda+1536xy-4096y^2-2048y^2\lambda$

$\Delta=-256x^2-128x^2\lambda+1536xy-4096y^2-2048y^2\lambda$


$$(-4\sqrt{2},-\sqrt{2},-\frac{7}{2})$$

$\Delta=-256(-4\sqrt{2})^2-128(-4\sqrt{2})^2(-\frac{7}{2})+1536(-4\sqrt{2})(-\sqrt{2})-4096(-\sqrt{2})^2-2048(-\sqrt{2})^2(-\frac{7}{2})=-256*32+128*16*7+1536*8-4096*2+2048*7=24576 > 0$

$$(-4\sqrt{2},\sqrt{2},-\frac{1}{2})$$ 

$\Delta=-256(-4\sqrt{2})^2-128(-4\sqrt{2})^2(-\frac{1}{2})+1536(-4\sqrt{2})(\sqrt{2})-4096(\sqrt{2})^2-2048(\sqrt{2})^2(-\frac{1}{2})=-256*32+128*16-1536*8-4096*2+2048=-24576 < 0$

$$(4\sqrt{2},-\sqrt{2},-\frac{1}{2})$$

$\Delta=-256(4\sqrt{2})^2-128(4\sqrt{2})^2(-\frac{1}{2})+1536(4\sqrt{2})(-\sqrt{2})-4096(-\sqrt{2})^2-2048(-\sqrt{2})^2(-\frac{1}{2})=-256*32+128*16-1536*8-4096*2+2048=-24576 < 0$

$$(4\sqrt{2},\sqrt{2},-\frac{7}{2})$$ 

$\Delta=-256(4\sqrt{2})^2-128(4\sqrt{2})^2(-\frac{7}{2})+1536(4\sqrt{2})(\sqrt{2})-4096(\sqrt{2})^2-2048(\sqrt{2})^2(-\frac{7}{2})=-256*32+128*16*7+1536*8-4096*2+2048*7=24576 > 0$

**Ответ:** 

${\color {red}{(-4\sqrt{2},-\sqrt{2},-\frac{7}{2}), (4\sqrt{2},\sqrt{2},-\frac{7}{2}) - точки\,\,максимума}}$

${\color {red}{(-4\sqrt{2},\sqrt{2},-\frac{1}{2}), (4\sqrt{2},-\sqrt{2},-\frac{1}{2}) - точки\,\,минимума }}$

```python
import numpy as np
from scipy.optimize import fsolve, broyden2
import math

def equations(p):
    x, y, z = p
    return (4*x+12*y+z*2*x,
            12*x+64*y+z*32*y,
            x**2+16*y**2-64)

solv = set()
for x in np.arange(-5, 5, 0.5):
  for y in np.arange(-5, 5, 0.5):
    for z in np.arange(-5, 5, 0.5):
          (x, y, z), info, ier, mesg =  fsolve(equations, (x, y, z), full_output=True)
          if (ier == 1): solv.add((round(x, 6), round(y, 6), round(z, 6)))

print(solv)
```

**__5*.__ Даны две функции $y_{1}=f(x)$ и $y_{2}=g(x)$. Известно, что:**

$f'(x)=\frac{1}{2}g(x)$

$g'(x)=2-2f(x)$

$f(0)=0$

$g(0)=1$

Восстановить функции $y_{1}=f(x)$ и $y_{2}=g(x)$ с помощью формулы:

$f(x+\Delta x) \approx f(x)+f'(x) \cdot \Delta x$

на участке $[0, 20]$

В качестве решения построить график этих функций в одной системе координат.


$f'(0)=\frac{1}{2}g(0)=\frac{1}{2}$

$g'(0)=2-2f(0)=2$

$\Delta x = 1$

$f(0)=0$

$f(0+1=1) = f(0)+f'(0) * 1 \Rightarrow f(1) = \frac{1}{2}$ 

$f(1+1=2) = f(1)+f'(1) * 1\Rightarrow f(2)=\frac{1}{2}+\frac{1}{2}g(1)$
$f(1+1=2) = f(1)+f'(1) * 1\Rightarrow f(2)=\frac{1}{2}+\frac{1}{2}*3=2$

$f(2+1=3) = f(2)+f'(2) * 1\Rightarrow f(3)=\frac{1}{2}+\frac{1}{2}g(1)+\frac{1}{2}g(2)$
$f(2+1=3) = f(2)+f'(2) * 1\Rightarrow f(3)=2+\frac{1}{2}*(5-2*\frac{1}{2})=4$

$g(0)=1$

$g(0+1=1) = g(0)+g'(0) * 1 \Rightarrow g(1) = 1+2=3$ 

$g(1+1=2) = g(1)+g'(1) * 1 \Rightarrow g(2) = 3+2-2f(1)=5-2f(1)=4$ 

$g(2+1=3) = g(2)+g'(2) * 1 \Rightarrow g(3) = 5-2f(1)+2-2f(2)=7-2f(1)-2f(2)=2$

```python
from scipy.interpolate import Akima1DInterpolator, splrep, splev
import numpy as np
import matplotlib.pyplot as plt

delta_x = 0.002
fx_points = []
gx_points = []
x_points = []

# Соберем точки функций f(x) и g(x) на отрезке [0, 20]
for i in np.arange(0, 20, delta_x):
    if i == 0:
        x_points.append(i)        
        fx_0 = 0
        fx_points.append(fx_0)
        gx_0 = 1        
        gx_points.append(gx_0)     
    else:
        x_points.append(i)
        fx_i = fx_points[-1] + (1/2)*gx_points[-1]*delta_x
        fx_points.append(fx_i)
        gx_i = gx_points[-1] + (2 - 2*fx_points[-2])*delta_x
        gx_points.append(gx_i)

# plt.plot(x_points, fx_points, marker="o", ls="")
# plt.plot(x_points, gx_points, marker="o", ls="")

# Сплайн через splrep, splev
spl1 = splrep(x_points, fx_points)
def fx(k):
    return splev(k, spl1)

spl2 = splrep(x_points, gx_points)
def gx(a):
    return splev(a, spl2)

yi1 = fx(x_points)
plt.plot(x_points, yi1, label="Функция f(x)")

yi2 = gx(x_points)
plt.plot(x_points, yi2, label="Функция g(x)")

plt.legend()
plt.show()
```
